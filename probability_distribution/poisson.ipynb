```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Step 1: Set the average rate (λ, lambda)
lambda_rate = 5   # Avg number of calls per hour

# Step 2: Create a range of possible call counts (0 to 15 calls)
x = [i for i in range(0, 16)]
print(x)

# Step 3: Calculate probability of each count
probabilities = poisson.pmf(x, mu=lambda_rate)

# Step 4: Plot it
plt.bar(x, probabilities, color='green')
plt.title("Poisson Distribution: Avg 5 calls/hour")
plt.xlabel("Number of calls in 1 hour")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```
