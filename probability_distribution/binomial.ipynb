```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n = 10       # Number of coin flips
p = 0.5      # Probability of heads

#  outcomes (0 to 10 heads)
x = np.arange(0, n+1)

# probability of each outcome using binomial dist
probabilities = binom.pmf(x, n, p)
print(probabilities)

#Plot it
plt.bar(x, probabilities)
plt.title("Binomial Distribution: 10 coin flips")
plt.xlabel("Number of heads")
plt.ylabel("Probability")
plt.grid(True)
plt.show()
```

**The Binomial Distribution answers questions like:“If I repeat an experiment n times, and each time there is a success or failure (like heads or tails), what’s the chance of getting k successes?”**

The variance of a Binomial Distribution is:
Variance=n×p×(1−p)
