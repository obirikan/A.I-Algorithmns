```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define mean and standard deviation
mu = 70     # average score
sigma = 10  # spread of scores

# Create a range of scores from 40 to 100
x = np.linspace(40, 100, 500)

# Calculate the probability density function (PDF)
pdf = norm.pdf(x, loc=mu, scale=sigma)

cdf = norm.cdf(80, loc=mu, scale=sigma)
print(cdf)

# Plot the curve
plt.plot(x, pdf, color='blue')
plt.title("Normal Distribution (mean=70, std=10)")
plt.xlabel("Score")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

## Probability Density Function

PDF tells you how dense the probability is around a specific value.
Not the probability of the exact value (since it's continuous), but how likely values are near it.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 70
std_dev = 10

x = np.linspace(40, 100, 500)
pdf = norm.pdf(x, mean, std_dev)

plt.plot(x, pdf, label="Normal Distribution")

# between 60 and 80
x_fill = np.linspace(60, 80, 500)
y_fill = norm.pdf(x_fill, mean, std_dev)
plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.5, label="P(60 ≤ X ≤ 80)")

# labels
plt.title("Normal Distribution with μ=70, σ=10")
plt.xlabel("Score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()

# Calculate probability
p = norm.cdf(80, mean, std_dev) - norm.cdf(60, mean, std_dev)
print(f"Probability of scoring between 60 and 80: {p:.4f}")
```

## Cumulative Distribution Function
norm.cdf() is used to calculate the cumulative probability up to a certain value.

It answers the question:
“What is the probability that a value from this distribution is less than or equal to X?”
