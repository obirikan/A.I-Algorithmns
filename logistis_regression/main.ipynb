```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Probability values from 0.01 to 0.99
p_values = np.linspace(0.01, 0.99, 100)

# Odds = p / (1 - p)
odds = p_values / (1 - p_values)

# Logit = log(odds)
logit = np.log(odds)

# 4. Visualize Odds and Logit
plt.figure(figsize=(15, 5))

# Odds vs Probability
plt.subplot(1, 2, 1)
plt.plot(p_values, odds, color='purple')
plt.title("Odds = p / (1 - p)")
plt.xlabel("Probability (p)")
plt.ylabel("Odds")
plt.grid(True)

# Logit vs Probability
plt.subplot(1, 2, 2)
plt.plot(p_values, logit, color='orange')
plt.title("Logit = log(p / (1 - p))")
plt.xlabel("Probability (p)")
plt.ylabel("Logit")
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Logistic Function Plot

```python
## logistic function
def logistic(x, b0=0, b1=1):
    return 1 / (1 + np.exp(-(b0 + b1 * x)))

x = np.linspace(-10, 10, 100)
p = logistic(x)

plt.figure(figsize=(7, 5))
plt.plot(x, p, color='green')
plt.title("Logistic Function: p(X) = 1 / (1 + e^-(b0 + b1*X))")
plt.xlabel("X (score, input feature)")
plt.ylabel("Probability (p)")
plt.grid(True)
plt.axhline(0.5, color='gray', linestyle='--', label='p = 0.5')
plt.legend()
plt.show()
```
